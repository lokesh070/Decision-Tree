```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
df = pd.read_csv("C:\\Users\\cheta\\Downloads\\adult_dataset.csv")
```

```python
df.head(4)
```

```python
df.shape
```

```python
df[df['workclass'] == '?'].shape
```

```python
df = df[df['workclass']!='?']
df.head(4)
```

```python
df_categorical = df.select_dtypes(include=['object'])
df_categorical.apply(lambda x: x=='?',axis=1).sum()
```

```python
df = df[df['occupation'] != "?"]
```

```python
df = df[df['native.country'] != "?"]
```

```python
df.isna().sum()
```

```python
df.occupation.value_counts()
```

```python
df_categorical = df.select_dtypes(include=['object'])
df_categorical.head()
```

```python
le = LabelEncoder()
df_categorical = df_categorical.apply(le.fit_transform)
df_categorical.head(4)
```

```python
df = df.drop(df_categorical.columns, axis=1)
```

```python
df = pd.concat([df,df_categorical],axis=1)
df.head(4)
```

```python
df['income'] = df['income'].astype('category')
```

```python
df.income.value_counts()
```

```python
X = df.drop(['fnlwgt','income'],axis=1)
y = df['income']
```

```python
X.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.30, 
                                                    random_state = 99)
```

```python
X_train.head(4)
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred_default = dt.predict(X_test)
```

```python
y_pred_train = dt.predict(X_train)
```

```python
print(classification_report(y_test,y_pred_default))
```

```python
print(classification_report(y_train,y_pred_train))
```

```python
dt = DecisionTreeClassifier(max_depth = 5)
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred_default = dt.predict(X_test)
```

```python
y_pred_train = dt.predict(X_train)
```

```python
print(classification_report(y_test,y_pred_default))
```

```python
print(classification_report(y_train,y_pred_train))
```
